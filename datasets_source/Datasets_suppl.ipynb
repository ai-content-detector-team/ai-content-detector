```python

# Liste des DataFrames à concaténer
dataframes = [df, df2, df3, df4, df5, df6, df7, df8]

# Concaténer toutes les colonnes 1 (colonne texte)
all_texts = pd.concat([d.iloc[:, 0] for d in dataframes], ignore_index=True)

# Concaténer toutes les colonnes 2 (colonne label)
all_labels = pd.concat([d.iloc[:, 1] for d in dataframes], ignore_index=True)

# Créer le DataFrame final avec les deux colonnes
df_final = pd.DataFrame({
    "text": all_texts,
    "label": all_labels
})

# Afficher un aperçu
print(df_final.head())
print(df_final["label"].value_counts())
```

```python
# Tout convertir en string au cas où il y a des entiers
df_final["label"] = df_final["label"].astype(str).str.lower()

# Remplacer toutes les variantes par les bons labels
df_final["label"] = df_final["label"].replace({
    "ai": "IA",
    "ia": "IA",
    "0": "IA",
    "human": "Human",
    "1": "Human"
})
```
